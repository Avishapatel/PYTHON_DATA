```python

#Write a function that replace all occurrences of "Java" with "Python" in above file
with open("practice.txt","r") as f:
    print("Original data")
    data=f.read()
    print(data)
new_data=data.replace("Java","Python")
print("new data")
print(new_data)
with open("practice.txt","w") as f:
    f.write(new_data)

```

```python
#find first "learining" word in which line. if "learning" word is not found in file then print -1

def check_line():
    line_no=1
    i=True
    with open("practice.txt","r") as f:
        while i:
            data=f.readline()
            if "learning" in data:
                return line_no
                break
            line_no+1
    return -1
ans=check_line()
print(ans)       

```

```python
#from a file containing numbers sepreted by comma,print the count of even number
with open("practice2.txt","r") as f:
    data=f.read()
a=data.split(",")
print(a)
count=0
for i in a:
    if int(i)%2==0:
        #print(i,"is even")
        count=count+1
print(count)
```
