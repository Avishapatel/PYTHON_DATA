```python


# Write a program to check whether a given character is a vowel, consonant, digit, or special character.
n1=input("Enter data = ")
if len(n1)!=1:
    print("Enter only one character data")
else:
    if n1.isalpha():
        if n1 in 'aeiouAEIOU':
            print(n1,"is VOWEL")
        else:
            print(n1,"is CONSONANT")
    elif n1.isdigit():
        print(n1,"is DIGIT")
    else:
        print(n1,"is SPECIAL CHARACTER")
```

```python
# Classify a character entered by the user as uppercase, lowercase, digit, or special symbol.
n2=input("Enter data = ")
if len(n2)!=1:
    print("Enter only one character data")
else:
    if n2.isalpha():
        if n2.islower():
            print(n2,"is lowercase character")
        else:
            print(n2,"is uppercase character")
    elif n2.isdigit():
        print(n2,"is DIGIT")
    else:
        print(n2,"is SPECIAL CHARACTER")
```

```python
# Find the second largest among three numbers using only if..else.
n1=int(input("Enter number 1="))
n2=int(input("Enter number 2="))
n3=int(input("Enter number 3="))
if n1<n2<n3 or n3<n2<n1:
    print(f"{n2} is second largest number")
elif n2<n3<n1 or n1<n3<n2:
    print(f"{n3} is second largest number")
elif n3<n1<n2 or n2<n1<n3:
    print(f"{n1} is second largest number")
else:
    print("Invalid")
```

```python
# A shop gives discounts: If the bill is more than 5000 → 20% discount, if between 2000–5000 → 10% discount, else no discount. Calculate the final price.
bill=float(input("Enter your bill amount")) 
if bill>5000:
    print("you get 20% discount")
    final_bill=(80*bill)/100
    print("Your final bill amount is = ",final_bill)
elif bill>=2000 and bill<=5000:
    print("you get 10% discount")
    final_bill=(90*bill)/100
    print("Your final bill amount is = ",final_bill)
else:
    print("no discount")
    print("your bill amount is = ",bill)
```

```python
# Take an integer input and check whether it is even or odd, and if even, also check if it is divisible by 4.
number=int(input("Enter value"))
if number%2==0:
    print(f"{number} is even number")
    if number%4==0:
        print(f"{number} is divisable by 4")
    else:
        print(f"{number} is not divisable by 4")
else:
    print(f"{number} is odd number")
```

```python
# Take marks of 5 subjects from the user, calculate the percentage and print the grade using if..else (A, B, C, Fail).
english=int(input("Enter marks of English="))
hindi=int(input("Enter marks of hindi="))
guj=int(input("Enter marks of gujarati="))
sci=int(input("Enter marks of science="))
maths=int(input("Enter marks of maths="))
per=((english+hindi+guj+sci+maths)*100)/500
print("Percentage=",per)

if per<=100 and per>90:
    print("A Grade")
elif per<=90 and per>75:
    print("B Grade")
elif per<=75 and per>50:
    print("C Grade")
else:
    print("Fail")





```
