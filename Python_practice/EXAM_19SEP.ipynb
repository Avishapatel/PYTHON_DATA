```python
# Q3. Write a Python program to demonstrate the use of the super() function in inheritance.

class Vehicle:
    def __init__(self,name):
        self.name=name
    def vehicle_info(self):
        print("vehicle details")
class Car(Vehicle):
    def __init__(self,name,brand):
        self.name=name
        self.brand=brand
    def car_info(self):
        super().vehicle_info()
        print("Car info")
c=Car("Baleno","Suzuki")
c.car_info()
```

```python
# Q4. Write a Python program to check if a given string is a palindrome.. (5 Marks)

str1=input("Enter string = ")
rev=str1[::-1]
if rev==str1:
    print(f"Given string ''{str1}'' is palindrome")
else:
    print(f"Given string ''{str1}'' is not palindrome")
```

```python
# Q8. What is the difference between a class and an object in Python? Give an example.(5 Marks)
# ans : class is blue print and object is a instance of class
# class is created by class keyword and access this class by creating an object
# syntax :: object_name=class_name(attributes) 

class Student:
    def __init__(self,name,std,subject,div):
        self.name=name
        self.std=std
        self.subject=subject
        self.div=div
    def Student_info(self):
        print("Name = ",self.name)
        print("Standard = ",self.std)
        print("Subject = ",self.subject)
        print("Division = ",self.div)
Stu=Student("Yashvi","11","Science","A")
Stu.Student_info()
```

```python
# Q9. Explain method overriding in Python with an example.(5 Marks)

# ans : method overriding means child class have same method or function of parents class with same signature

class vehicle:
    def __init__(self,name,color,brand):
        self.name=name
        self.color=color
        self.brand=brand
    def information(self):
        print("Name of Vehicle = ",self.name)
        print("Color of Vehicle = ",self.color)
        print("Brand of Vehicle = ",self.brand)
class Car:
    def __init__(self,name,color,brand):
        self.name=name
        self.color=color
        self.brand=brand
    def information(self):
        print("Name of Car = ",self.name)
        print("Color of Car = ",self.color)
        print("Brand of Car = ",self.brand)
C=Car("Swift","White","Suzuki")
C.information()
V=vehicle("Truck","Red","KIA")
V.information()
```
