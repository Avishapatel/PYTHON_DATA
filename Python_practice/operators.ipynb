```python
#Arithmatic operators
a=30
b=12
print(a+b,a-b,a*b,a/b,a%b,a//b,a**b)
```

```python
#assignment operators
a=40
b=20
print(a)
print(b)
a+=2
print(a)
a-=3
print(a)
b*=3
print(b)
b/=2
print(b)


```

```python
#Relational or Comparison Operators
a=78
b=9
print(a>b,a<b,a>=b,a<=b,a==b,a!=b)

```

```python
#Logical Operators
a=99
b=78
c=76
if a>b and a>c:
    print("a is grater")
else:
    print("nothing")

if b<a or b<c:
    print("welcome")
```

```python
#Membership operators

a=["avisha","hardik","parth"]
print('avisha' in a)
print('parth' not in a)
```

```python
#Identity Operators
a=["avisha","hardik","parth"]
b=["avisha","hardik","parth"]
print(id(a))
print(id(b))
print(a is b)
print(a is not b)
```
