```python
# Task 2: Advanced FizzBuzz Variant

# Loop from 1 to 100:

# If divisible by 3 and 5 → print "FizzBuzz"

# If divisible by 7 → print "Lucky"

# If divisible by 11 → print "Boom"

# Else → print the number

for i in range(1,100+1):
    if i%3==0 and i%5==0:
        print("Fizzbuzz")
    elif i%7==0:
        print("Lucky")
    elif i%11==0:
        print("Boom")
    else:
        print(i)
```

```python
# Task 3: Student Marks Analyzer

# Take marks of n students in 3 subjects.

# Use a for loop to input marks.

# Use if-else to decide Pass/Fail (pass if avg ≥ 40 and no subject < 35).

# Print highest scorer and class average at the end.

student=int(input("enter number of student"))
averagedata=[]
student_name=[]
for i in range(1,student+1):
    stu_name=input("Enter student name = ")
    print("Result of ",stu_name)
    student_name.append(stu_name)
    s1=int(input("Enter marks of Hindi = ")) 
    s2=int(input("Enter marks of English = "))
    s3=int(input("Enter marks of Gujarati = "))

    avg=(s1+s2+s3)/3
    averagedata.append(avg)
    if avg>=40:
        print("PASS")
    else:
        print("FAIL")
print(averagedata)
print(student_name)
result=max(averagedata)
r=averagedata.index(result)
print(r)
toper_name=student_name[r]
print("Higest scorer is",toper_name)
print("Higest score of class is ",result)
```

```python
# Task 4 : Armstrong Number Finder

# Loop through numbers 100–999.

# Use if-else to check if the sum of cubes of digits equals the number.

# Print all Armstrong numbers.
armstrong=[]
not_arm=[]
for i in range(100,1000):
    s=str(i)
    l=list(s) 
    # print(l)
    sum=(int(l[0])**3)+(int(l[1])**3)+(int(l[2])**3)
    # print("sum =",sum)
    if sum==i:
        # print(f"{i} number is Armstrong number")
        armstrong.append(i)
    else:
        # print(f"{i} number not Armstrong number")
        not_arm.append(i)
print("Total Armstrong number is",armstrong)
print("not Armstrong",not_arm)
```

```python
# Task : Armstrong Number Finder

# check user enter number is armstrong or not.

# Use if-else to check if the sum of cubes of digits equals the number.

# Print all Armstrong numbers.

nn=int(input("Enter value = "))
def armstrong(arm):
    data=str(arm)
    length=len(data)
    # print(length)
    d=list(data)
    # print(d)
    sum=0
    for i in range(length):
        # print(d[i])
        r=int(d[i])**length
        # print(r)
        sum=sum+r
    return sum
result=armstrong(nn)
print(result)
if nn==result:
    print(f"your value {nn} is armstrong")
else:
    print(f"your value {nn} is not armstrong")
```

```python
# Task 5: Bank ATM Simulation

# User enters balance and a series of transactions (deposit/withdraw).

# Use a for loop to process transactions.

# Use if-else to validate withdrawal (balance >= amount).

# Print final balance after all transactions.

balance=int(input("Enter total balance = "))
transaction=int(input("Enter number of transactions series = "))
for i in range(transaction):
    withdrawal_amount=int(input(f"Enter amount for withdrawal = "))
    if balance>=withdrawal_amount:
        balance=balance-withdrawal_amount
        print("your current balance is ",balance)
    else:
        print("your current balance is less than withdrawal amount")
        break
print("your Final Amount in account is ",balance)


```
