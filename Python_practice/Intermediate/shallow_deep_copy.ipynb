```python
import copy
a=[[1,2,3,4,5],[0,9,8,7,6],[11,12,13,14]]
shallow_copy=copy.copy(a)
a[1][3]=20
a[0][1]=10
a[2][2]=30
print("Original copy = ",a)
print("Shallow copy = ",shallow_copy)
print(a is shallow_copy)       #different outer list
print(a[1] is shallow_copy[1])   #same inner list
```

```python
import copy
a=[[1,2,3,4,5],[0,9,8,7,6],[11,12,13,14]]
deep_copy=copy.deepcopy(a)
a[1][3]=20
a[0][1]=10
a[2][2]=30
print("Original copy = ",a)
print("Shallow copy = ",deep_copy)
print(a is deep_copy)       #different outer list
print(a[1] is deep_copy[1])   #different inner loop
```
