```python
ans=lambda x,y:x*x*y
print(ans(20,34))
```

```python
a=lambda c: "Even" if c%2==0 else "odd"
print(a(36))
```

```python
pi=3.14
n=int(input("enter value of r "))
ans=lambda r:pi*r*r
print(ans(n))
```

```python
p=[24,65,45,89]
q=list(filter(lambda x: "is Even" if x%2==0 else'',p))
print(q)
```

```python
p=[24,65,45,89]
q=list(map(lambda x:f"{x} is Even" if x%2==0 else f"{x} is odd",p))
print(q)
```
