```python
#calculate electricity bill.

unit=int(input("Enter total units = "))
if unit>=0 and unit<=100:
    bill=unit*5
    print("Total bill Amount is",bill)
elif unit>=101 and unit<=200:
    bill=unit*7
    print("Total bill Amount is",bill)
elif unit>200:
    bill=unit*10
    print("Total bill Amount is",bill)
else:
    print("Invalid Data")

```

```python
#GCD of two number
num1=int(input("Enter value 1 = "))
num2=int(input("Enter value 2 = "))
result1=[]
result2=[]
for i in range(2,num1):
    if num1%i==0:
        result1.append(i)

for j in range(2,num2):
    if num2%j==0:
        result2.append(j)
# print(result1)
# print(result2)
c=set(result1) & set(result2)
# print(c)
result=list(c)
gcd=max(result)
print(f"GCD OF {num1} AND {num2} is {gcd}")
```
