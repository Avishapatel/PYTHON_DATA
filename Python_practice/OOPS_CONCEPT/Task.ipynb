```python
# Create a Class and Object

# Make a Car class with attributes like brand, model, and year.

# Create multiple objects and print their details.

# Class with Methods

# Add methods like start() and stop() in the Car class.

# Show how objects call these methods.


class Car:
    brand="Suzuki"
    model="Swift"
    year=2002
    def start(self):
        print("Start the car")
    def stop(self):
        print("Stop the car")
    
c=Car()
print(c.brand )
print(c.model)
print(c.year)  
c.start()
c.stop()    
```

```python
# Single Inheritance

# Class Employee (base) → Class Manager (derived).

# Show inheritance of methods.

class Employee:
    def empl_info(self):
        print("my name is hardik")
class Manager(Employee):
    def manager_info(self):
        print("i am manager of KIARA company")
man=Manager()
man.empl_info()
man.manager_info()
```

```python
# Multiple Inheritance

# Create Teacher and Researcher classes.

# Inherit them into Professor class.

class Teacher:
    def teacher_info(self):
        print("I am Teacher")
class Researcher:
    def researcher_info(self):
        print("I am Researcher")
class Professor(Teacher,Researcher):
    def Professor_info(self):
        self.teacher_info()
        self.researcher_info()
pro=Professor()
pro.Professor_info()
```

```python

# Multilevel Inheritance

# Animal → Mammal → Dog.

# Each class should have its own method.

class Animal:
    def animal_info(self):
        print("ANIMAL")
class Mammal(Animal):
    def mammal_info(self):
        print("MEMMAL")
        self.animal_info()
class Dog(Mammal):
    def dog_info(self):
        print("DOG")
        self.mammal_info()
d=Dog()
d.dog_info()      

```
