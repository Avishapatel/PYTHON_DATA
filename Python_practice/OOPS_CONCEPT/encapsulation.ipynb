```python
class ATM:
    def __init__(self,balance=50000,pin=None):
        self.__balance=balance
        self.__pin=pin
        self.details()
    def details(self):
        inp=int(input("Press 1 for pin creation \n press 2 for balance check \n press 3 for withdrawal \n press 4 for deposite "))
        if inp==1:
            self.pin_creation()
        elif inp==2:
            self.balance_check()
        elif inp==3:
            self.withdrawal()
        elif inp==4:
            self.deposite()
        else:
            print("invalid data")
    def pin_creation(self):
        self.__pin=int(input("enter your pin = "))
        print("your pin is",self.__pin)
    def balance_check(self):
        temp=int(input("Enter your pin"))
        if temp==self.__pin:
            print("your_balance is ",self.__balance)
        else:
            print("Invalid pin")
    def deposite(self):
        temp=int(input("enter your pin"))   
        if temp==self.__pin:
            dep=int(input("Enter your deposite ammount ")) 
            self.__balance=self.__balance+dep     
            print("your total balance is",self.__balance)   
        else:
            print("invalid pin")
    def withdrawal(self):
        temp=int(input("enter your pin = "))
        if temp==self.__pin:
            wid=int(input("enter withdrawal amount ="))
            self.__balance=self.__balance-wid
            print("your total balance is",self.__balance)
        else:
            print("Invalid pin")  

s=ATM() 
s.details()  
# print(s.__pin)        #private member
```
