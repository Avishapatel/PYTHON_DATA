```python
 
 

 
 
 
 
 
from abc import ABC, abstractmethod 
  
# Abstract class 
class Animal(ABC): 
     
    @abstractmethod 
    def make_sound(self): 
        pass  # No implementation here 
  
# Child classes must implement the abstract method 
class Dog(Animal): 
    def make_sound(self): 
        print("Bark") 
  
class Cat(Animal): 
    def make_sound(self): 
        print("Meow") 
  
# animal = Animal()  # ❌ This will give an error — can't instantiate abstract class 
  
dog = Dog() 
cat = Cat() 
  
dog.make_sound()  
cat.make_sound()  
 
```
