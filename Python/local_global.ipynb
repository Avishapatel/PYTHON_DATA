```python
<!-- A local variable is created inside a function.

It exists only while the function is running.

You cannot access it outside the function. -->

 <!-- global Keyword

To actually modify a global variable inside a function, use the global keyword. -->

<!-- Local variable → declared inside a function, accessible only inside.

Global variable → declared outside functions, accessible everywhere. -->
```

```python
x=10
def myfun():
    global x
    x=10
myfun()
print(x)
```

```python
y=20
def myfun():
    print(y)
myfun()
print(y)
```
