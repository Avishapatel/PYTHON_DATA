```python
# Nested If:
# if statements inside if statements, this is called nested if statements.

n=int(input("Enter the number"))
if n>5:
    if n<20:
        print("between 5 to 20")
    else:
        print("20 is higher")
else:
    print("value is lower tahn 5")
```
