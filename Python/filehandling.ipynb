```python
# File Handling

# "r" → Read (default)

# "w" → Write (overwrites file if exists)

# "a" → Append (add new data at end)

# "b" → Binary mode (e.g., images, PDFs)

# "x" → Create new file, error if exists

# "r+" → Read & write
```

```python
# Read entire file
with open("example.txt", "r") as f:
    data = f.read()
    print(data)
```

```python
# Read line by line
with open("example.txt", "r") as f:
    for line in f:
        print(line.strip())
```

```python
# Overwrite file with new data
with open("example.txt", "w") as f:
    f.write("Hello, Python!\n")
    f.write("This is a new line.\n")
```

```python
# Read as list of lines
with open("example.txt", "r") as f:
    lines = f.readlines()
    print(lines)
```
