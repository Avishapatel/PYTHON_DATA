```python
i=True

while i:

    n1=int(input("Enter value "))
    if n1%2==0:
        print(n1,"is even number")
    else:
        print(n1,"is odd number")
    choice=input("Enter your choice yes or no =").strip().lower()
    if choice=="yes":
        i=True
    else:
        i=False
```

```python
j=True
while j:
    n=int(input("Enter year = "))
    if n%4==0 and n%100!=0 or n%400==0:
        print(n,"leap year")
    else:
        print(n,"not leap year")
    choice=input("enter yes or no ").strip().lower()
    if choice=="yes":
        j=True
    else:
        j=False
```

```python
print('''  Domino's MENU
            PIZZA = 400
            SANDWITCH = 200
            BURGER = 300
            FRENCHFRIES = 100  ''')
i=True
total=0
while i:
    item=input("Enter PIZZA,SANDWITCH,BURGER,FRENCHFRIES = ")
    qauntity=int(input("Enter no of quantity = "))
    if item=="PIZZA":
        total=total+(qauntity*400)
    elif item=="SANDWITCH":
        total=total+(qauntity*200)
    elif item=="BURGER":
        
        total=total+(qauntity*300)
    elif item=="FRENCHFRIES":
        total=total+(qauntity*100)
    else:
        print("ENTER valid data")
    
    ch=input("Enter your choice (yes/no) =").strip().lower()
    if ch=="yes":
        i=True
    else:
        i=False
print("Your total bill amount = ",total)
```

```python
import random
computer=random.randint(1,10)
i=True
while i:
    user=int(input("Enter number "))
    if user<computer:
        print("Enter up number")
    elif user>computer:
        print("Enter lower number")
    elif user==computer:
        print("Both are same")
        i=False
    else:
        print("invalid number")
```
