```python
# List slicing is a technique used to extract a subset (slice) of elements from a list by specifying a range of indices.
# list[start : stop : step]
# start → Index to begin the slice (inclusive)

# stop → Index to end the slice (exclusive)

# step → (Optional) How many steps to skip


```

```python
a="code vibe innovation"
# print(a[0:4])
# print(a[5:9])
# print(a[0:20:2])  # step 
print(a[5:]) # stating position only 
print(a[:10]) # Ending position only 
```
