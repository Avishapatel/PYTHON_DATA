```python
# * The process of inheriting the properties of the parent class into a child class is called inheritance. The existing class is     called a base class or parent class and the new class is called a subclass or child class or derived class.
# * In inheritance, the child class acquires all the data members, properties, and functions from the parent class. Also, a child   class can also provide its specific implementation to the methods of the parent class.
# * For example, In the real world, Car is a sub-class of a Vehicle class. We can create a Car by inheriting the properties of a     Vehicle such as Wheels, Colors, Fuel tank, engine, and add extra properties in Car as required.
```

```python
# * Syntax :
# --> class BaseClass:
#         Body of base class
#     class DerivedClass(BaseClass):
#         Body of derived class

# * Types Of Inheritance :-  
# --> based upon the number of child and parent classes involved, there are five types of inheritance.
#       1. Single inheritance
#       2. Multiple Inheritance
#       3. Multilevel inheritance
#       4. Hierarchical Inheritance
#       5. Hybrid Inheritance

```

```python
# * In single inheritance, a child class inherits from a single-parent class. Here is one child class and one parent class.


class Vehicle:
    def vehicle_info(self):
        print("Vehicle Information")
class  Car(Vehicle):
    def car_info(self):
        print("Car_Information")

C=Car()
C.vehicle_info()
C.car_info()

```

```python
class student:
    def student_data(self):
        print("hello student")
class std10_student(student):
    def std10_data(self):
        print("class 10th")
ten=std10_student()
ten.student_data()
ten.std10_data()
```

```python
# * In multiple inheritance, one child class can inherit from multiple parent classes. So here is one child class and multiple parent classes.

class Person:
    def personal_info(self,name):
        print("Name",name)

class Company:
    def Company_info(self,c_name):
        print("Company_Name",c_name)

class Employee(Person,Company):
    def Employee_Info(self,salary):
        print("Salary",salary)

Emp=Employee()
Emp.personal_info('Aadhya')
Emp.Company_info('Google')
Emp.Employee_Info(250000)
```

```python
class Person:
    def person_details(self,name,age):
        print("Name =",name)
        print("Age = ",age)
class Degree:
    def degree_details(self,degree):
        print("Degree = ",degree)

class student(Person,Degree):
    def student_details(self):
        print("hello student")
s=student()
s.person_details("parth",25)
s.degree_details("B.com")
s.student_details()
```

```python
# * In multilevel inheritance, a class inherits from a child class or derived class. Suppose three classes A, B, C. A is the         superclass, B is the child class of A, C is the child class of B. In other words, we can say a chain of classes is called       multilevel inheritance.

class Vehicle:
    def Vehicle_Info(Self):
        print("Vehicle_Information")

class Car(Vehicle):
    def Car_Info(self):
        print("Car_Information")

class Sports_car(Car):
    def Sports_Car(self):
        print("Sports_Car Info")

s=Sports_car()
s.Sports_Car()
s.Vehicle_Info()
s.Car_Info()
```

```python
class Grandparents:
    def Grandparents_details(self,mname,fname):
        print("GrandMother name = ",mname)
        print("Grandfathername = ",fname)
class Parents(Grandparents):
    def parents_details(self,mname,fname):
        print("Mother name = ",mname)
        print("fathername = ",fname)
class Child(Parents):
    def child_details(self,name):
        print("Child name = ",name)

ch=Child()
ch.child_details("Avisha")
ch.parents_details("Dipikaben","Mukeshbhai")
ch.Grandparents_details("Ambaben","Ambalal")
```
