```python
# Abstraction is one of the core concepts of OOP (Object-Oriented Programming).
# It means hiding implementation details and showing only the essential features to the user.
# In Python, abstraction is achieved using abstract classes and abstract methods, which are defined in the abc (Abstract Base Class) module.

# An abstract class cannot be instantiated (we cannot create objects directly).

# An abstract method is a method declared, but it must be implemented in the child class.

```

```python
from abc import ABC, abstractmethod
class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass
    def stop(self):
        pass
class bike(Vehicle):
    def start(self):
        print("start bike")
        pass
    def stop(self):
        print("stop bike")
        pass
class Car(Vehicle):
    def start(self):
        print("start car")
        pass
    def stop(self):
        print("stop car")
        pass
b=bike()
b.start()
b.stop()
c=Car()
c.start()
c.stop()
```

```python
from abc import ABC,abstractmethod
class RBI(ABC):
    @abstractmethod
    def get_interest_rate(self):
        pass
    @abstractmethod
    def get_min_balance(self):
        pass
    @abstractmethod
    def get_loan(self,amount):
        pass

class SBI(RBI):
    def get_interest_rate(self):
        return 6.5

    def get_min_balance(self):
        return 1000

    def get_loan(self, amount):
        print(f"SBI provides a loan of ₹{amount} at {self.get_interest_rate()}% interest.")

class HDFC(RBI):
    def get_interest_rate(self):
        return 7.2

    def get_min_balance(self):
        return 5000

    def get_loan(self, amount):
        print(f"HDFC provides a loan of ₹{amount} at {self.get_interest_rate()}% interest.")

bank1 = SBI()
bank2 = HDFC()
print("----- SBI -----")
print("Min Balance:", bank1.get_min_balance())
bank1.get_loan(200000)


print("----- HDFC -----")
print("Min Balance:", bank2.get_min_balance())
bank2.get_loan(300000)





```
