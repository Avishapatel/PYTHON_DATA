```python
# * The word polymorphism is taken from the Greek words poly (many) and morphism (forms). It means a method can process objects     differently depending on the class type or data type.
# * Polymorphism in Python is the ability of an object to take many forms. In simple words, polymorphism allows us to perform the   same action in many different ways.
```

```python
# * In the method overriding, methods or functions must have the same name and same signatures



class Vehicle:
    def __init__(self,color,price,name):
        self.color=color
        self.price=price
        self.name=name
    def show(self):
        print("Details = ",self.name,self.price,self.color)
    def change_giyer(self):
        print("vihicle change 6 giyer")
    def maximum_speed(self):
        print("maximum speed 180")
class car(Vehicle):
    def __init__(self,color,price,name):
        self.color=color
        self.price=price
        self.name=name
    def show(self):
        print("car Details = ",self.name,self.price,self.color)
    def change_giyer(self):
        print("car change 6 giyer")
    def maximum_speed(self):
        print("maximum speed of car is 180")
c=car("RED",700000,"BALENO")
c.show()
c.change_giyer()
c.maximum_speed()
V=Vehicle("black",800000,"swift")
V.show()
V.change_giyer()
V.maximum_speed()
```

```python
# * In Python, we can change the default behavior of the built-in functions. For example, we can change or extend the built-in       functions such as len(), abs(), or divmod() by redefining them in our class.



class Shopping:
    def __init__(Self,basket,buyer):
        Self.basket=basket
        Self.buyer=buyer


    def __len__(self):
        print('Length')
        count=len(self.basket)
        return count*3

sh=Shopping(['shoes','Dress'],'Aadhya')
print(len(sh))

```

```python
class Ferrari:
    def fuel_type(Self):
        print("Petrol")

    def max_speed(self):
        print('Max speed is 200')

class BMW:
    def fuel_type(Self):
        print("Petrol")

    def max_speed(self):
        print('Max speed is 200')


f=Ferrari()
b=BMW()

for car in (f,b):
    car.fuel_type()
    car.max_speed()
```

```python
class Ferrari:
    def fuel_type(Self):
        print("Petrol")

    def max_speed(self):
        print('Max speed is 200')

class BMW:
    def fuel_type(Self):
        print("Petrol")

    def max_speed(self):
        print('Max speed is 200')

def car_details(obj):
    obj.fuel_type()
    obj.max_speed()
f=Ferrari()
b=BMW()

car_details(f)
car_details(b)
```

```python
# * In the method overloading, methods or functions must have the same name and different signature.

def product(a,b):
    p=a*b
    print(p)

def product(c,d,e):
    p=c-d
    print(p)

product(4,5)
product(4,5,6)
```
