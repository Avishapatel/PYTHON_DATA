```python
# Membership OPerator
# in =  Returns True if a sequence with the specified value is present in the object
# not in = Returns True if a sequence with the specified value is not present in the object

a=["Banana" , "Mango"]
print(type(a))
print("Banana" in a)
print("Mango"  not in a)

```
