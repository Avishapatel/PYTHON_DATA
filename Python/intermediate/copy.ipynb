```python
<!-- Shallow Copy  -->
# Definition: Creates a new object, but does not recursively copy nested objects.

# Instead, it just copies references of nested objects.

# So if you modify a nested object, changes will be reflected in both copies.
```

```python
import copy 

list=[[1,2,3],[4,5,6]]

shallow=copy.copy(list)

print(list)

list[1][1]=20

print("Original",list)
print("Shallow Copy",shallow)
```

```python
# Deep Copy

# definition: Creates a completely new object and recursively copies all nested objects.

# Changes in the original object do not affect the deep copy.

import copy

list=[[1,2,3],[4,5,6]]
deep=copy.deepcopy(list)

print(list)
list[1][1]=20

print("Original List",list)
print("Deep Copy",deep)
```

```python
# | Feature                                | Shallow Copy      | Deep Copy             |
# | -------------------------------------- | ----------------- | --------------------- |
# | Top-level copy                         | Yes               | Yes                   |
# | Nested objects                         | References shared | Completely new copies |
# | Changes in nested objects affect copy? | ✅ Yes             | ❌ No                  |
```
