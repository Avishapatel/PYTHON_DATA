```python
num=[1,2,3,4,5]
it=iter(num)
print(next(it))
print(next(it))
print(next(it))
print(next(it))
print(next(it))

```

```python
A generator is a special kind of iterator created with yield instead of return.

It generates values on the fly (lazy evaluation).

Saves memory (doesnâ€™t store the whole sequence at once).

Automatically implements the iterator protocol (__iter__, __next__).


```

```python
def my_generator():
    yield 1
    yield 2
    yield 3
v=my_generator()
print(next(v))
print(next(v))
print(next(v))
```

def my_gen():
    yield 1
    yield 2
    yield 3

g = my_gen()

print(next(g))  # 1
print(next(g))  # 2
print(next(g))  # 3
# next(g)  # Raises StopIteration


```python
# | Feature     | Iterator                                       | Generator                                     |
# | ----------- | ---------------------------------------------- | --------------------------------------------- |
# | Creation    | Defined using classes & `__iter__`, `__next__` | Defined using `def` + `yield` keyword         |
# | Memory      | May store all elements in memory               | Generates values one at a time (saves memory) |
# | Ease of use | More complex to write                          | Very easy with `yield`                        |
# | Example     | `iter([1,2,3])`                                | A function with `yield`                       |
```

```python
# Generator
def square(n):
    for i in range(n):
        yield i*i
for y in square(5):
        print(y)
```

```python
# A decorator is a function in Python that takes another function as an argument, 
# adds some extra functionality to it, and returns a new function.

def greet():
    return "Hello"
def upperc(fun):
    def wrapper():
        result=fun
        return result.upper()
    return wrapper
abc=upperc(greet())
print(abc())

```

```python
def uppercase(fun):
    def wrapper():
        result=fun()
        return result.upper()
    return wrapper
@uppercase
def read():
    return "hello"
print(read())
```
