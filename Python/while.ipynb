```python
# With the while loop we can execute a set of statements as long as a condition is true.

# Syntax :-
# --> while condition:
#         Block of statement(s)
```

```python
a=5
while a<10:
    print(a)
    a=a+1
```

```python
n = int(input("Please Enter Number (0 or 1 to stop): "))
while n>1 :
    if n%2==0:
        print("even number")
    else:
        print("Odd number")
    n = int(input("Please Enter Number (0 or 1 to stop): "))
    n=n+1
```

```python
size=5
for i in range(1,size+1):
    for j in range(size-i):
        print(' ',end=" ")
    for k in range(2*i-1):
        print('*',end=" ")
    print()
```

```python
for i in range(1, 6):              # Outer loop (rows)
    for j in range(1, i + 1):      # Inner loop (columns)
        print('*', end=" ")        # Print star in the same line
    print('')                      # Move to the next line
```

```python
for i in range(1,6):
    print(" " * ( 5- i), end="")   # spaces before stars
    print("* " * i)


# Row 1 (i=1): " " (4 spaces) + "* " → *

# Row 2 (i=2): " " (3 spaces) + "* * " → * *

# Row 3 (i=3): " " (2 spaces) + "* * * " → * * *

# Row 4 (i=4): " " (1 space) + "* * * * " → * * * *

# Row 5 (i=5): "" (0 spaces) + "* * * * * " → * * * * *


```
