```python
# Set are used to store multiple items in a single variable.
# created using "{}".
# mutable (Changeble)
# Duplication not Allow
# indexing not allow
# always give sort data
```

```python
s={1,2,3,4,5,2,2,2}
print(s)
```

```python
a={1,2,3,4}
b={2,3}
a.add(5)
print(a)
print(a.symmetric_difference(b))    #give different data from both set
```

```python
s={1,2,3,4}
s1={4,8,9,10,3}
print(s & s1)          #give same data from both set
print(s.union(s1))     #merge both set data
```
